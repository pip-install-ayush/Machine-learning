```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
gas= pd.read_csv("C:\\Users\\Ayush mishra\\Downloads\\gas_prices.csv")

plt.figure(figsize=(8,5))
plt.title('Gas Prices over Years', fontdict={'fontweight':'bold', 'fontsize':10})
# plt.plot(gas.Years, gas.['Canada'])
# plt.plot(gas.Year, gas.USA, '.-', label='USA')
# plt.plot(gas.Year, gas.Canada, '.-', label='Canada')
countries_to_look_at = ['Australia', 'USA', 'Canada', 'South Korea']
for country in gas:
    if country in countries_to_look_at:
        plt.plot(gas.Year, gas[country], marker='.', label = country)

plt.xticks(gas.Year[::3].tolist()+[2011])

plt.xlabel('Year')
plt.ylabel('US Dollars')

plt.xticks(gas.Year[::3])

plt.legend()
plt.show()
```
